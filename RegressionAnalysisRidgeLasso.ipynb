```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
```

```python
df1 = pd.read_csv('uber.csv')
```

```python
df1.drop(labels='Unnamed: 0',axis=1,inplace=True)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
# df.dropna(inplace=True)
```

```python
df1.fillna(value={'dropoff_longitude':df1['dropoff_longitude'].mean(),'dropoff_latitude':df1['dropoff_latitude'].mean()},inplace=True)
```

```python
df1.describe()
```

```python
df1['pickup_datetime'] =  pd.to_datetime(df1['pickup_datetime'])
```

```python
df1
```

```python
df1['hour'] = df1['pickup_datetime'].dt.hour
df1['dayofweek'] =  df1['pickup_datetime'].dt.dayofweek
```

```python
df = df1.drop(columns=['key','pickup_datetime'],axis=1)
df
```

```python
X = df.drop(columns=['fare_amount'],axis=1)
y = df['fare_amount']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)1
```

```python
Scaler = StandardScaler()
```

```python
X_Scaler_train = Scaler.fit_transform(X_train)
X_Scaler_test = Scaler.fit_transform(X_test)
```

```python
from scipy import stats
z_scores = stats.zscore(df['fare_amount'])
outliers = (z_scores > 3) | (z_scores < -3)
outlier_indices = df.index[outliers]

df = df.drop(outlier_indices)
```

```python
correlation = X_train.corr()
```

```python
correlation
```

```python
linear_model = LinearRegression()
linear_model.fit(X_Scaler_train,y_train)
```

```python
lasso = Lasso(alpha=1.0)
lasso.fit(X_Scaler_train,y_train)
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_Scaler_train,y_train)
```

```python
linear_prediction = linear_model.predict(X_Scaler_test)
linear_prediction
```

```python
lasso_prediction = lasso.predict(X_Scaler_test)
lasso_prediction
```

```python
ridge_prediction = ridge.predict(X_Scaler_test)
ridge_prediction
```

```python
r2_linear = r2_score(y_test, linear_prediction)
rmse_linear = sqrt(mean_squared_error(y_test, linear_prediction))
```

```python
r2_lasso = r2_score(y_test, lasso_prediction)
rmse_lasso = sqrt(mean_squared_error(y_test, lasso_prediction))
```

```python
r2_ridge = r2_score(y_test, ridge_prediction)
rmse_ridge = sqrt(mean_squared_error(y_test, ridge_prediction))
```

```python
print(f"Linear Regression - R2: {r2_linear}, RMSE: {rmse_linear}")
print(f"Ridge Regression - R2: {r2_ridge}, RMSE: {rmse_ridge}")
print(f"Lasso Regression - R2: {r2_lasso}, RMSE: {rmse_lasso}")
```
